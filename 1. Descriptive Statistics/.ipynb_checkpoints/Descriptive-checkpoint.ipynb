The notebook will print data from screener.in website output
It wull read Profit and Loss data, Balance sheet data, Cash flow data and Quarterly data and then 
generate plots for the same

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Tata Steel.xlsx"   # ensure exact filename

# ---------- 1. PROFIT & LOSS: denormalize + visuals ----------

pl_df = pd.read_excel(
    file_path,
    sheet_name="Data Sheet",
    skiprows=15,      # A16
    nrows=16,
    usecols="A:K"
)

pl_long = pl_df.melt(
    id_vars=["Report Date"],
    var_name="Date",
    value_name="Amount"
)
pl_long["Date"] = pd.to_datetime(pl_long["Date"])
pl_wide = pl_long.pivot_table(
    index="Date",
    columns="Report Date",
    values="Amount",
    aggfunc="sum"
).reset_index()
pl_wide["Year"] = pl_wide["Date"].dt.year

# 1a. Stacked bar: cost components
cost_components = [
    "Raw Material Cost",
    "Change in Inventory",
    "Power and Fuel",
    "Other Mfr. Exp",
    "Employee Cost",
    "Other Expenses"
]
cost_cols = [c for c in cost_components if c in pl_wide.columns]

fig, ax = plt.subplots(figsize=(12, 6))
pl_wide.set_index("Year")[cost_cols].plot(kind="bar", stacked=True, ax=ax)
ax.set_title("Cost Components Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(title="Components", loc="upper left", ncol=2, bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()

# 1b. Line chart: Sales and Net profit
sales_profit_cols = [c for c in ["Sales", "Net profit"] if c in pl_wide.columns]

fig, ax = plt.subplots(figsize=(12, 6))
pl_wide.set_index("Date")[sales_profit_cols].plot(kind="line", ax=ax)
ax.set_title("Sales and Profit Over Time")
ax.set_xlabel("Date")
ax.set_ylabel("Amount")
plt.xticks(rotation=45)
plt.legend(title="Sales / Profit", loc="upper left", ncol=1, bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()

# ---------- 2. BALANCE SHEET: denormalize + clustered assets/liabilities ----------

bs_df = pd.read_excel(
    file_path,
    sheet_name="Data Sheet",
    skiprows=55,      # A56
    nrows=15,
    usecols="A:K"
)

bs_long = bs_df.melt(
    id_vars=["Report Date"],
    var_name="Date",
    value_name="Amount"
)
bs_long["Date"] = pd.to_datetime(bs_long["Date"])
bs_wide = bs_long.pivot_table(
    index="Date",
    columns="Report Date",
    values="Amount",
    aggfunc="sum"
).reset_index()
bs_wide["Year"] = bs_wide["Date"].dt.year

asset_candidates = ["Cash", "Inventory", "Receivables", "Investment", "Net Block"]
liab_candidates  = ["Borrowings", "Other Liabilities", "Reserves", "Equity Share Capital"]

asset_cols = [c for c in asset_candidates if c in bs_wide.columns]
liab_cols  = [c for c in liab_candidates  if c in bs_wide.columns]

assets_wide = bs_wide[["Year"] + asset_cols].set_index("Year")
liabs_wide  = bs_wide[["Year"] + liab_cols].set_index("Year")

# 2a. Assets clustered columns
fig, ax = plt.subplots(figsize=(12, 6))
assets_wide.plot(kind="bar", ax=ax)
ax.set_title("Assets Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(title="Assets", loc="upper left", ncol=2, bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()

# 2b. Liabilities clustered columns
fig, ax = plt.subplots(figsize=(12, 6))
liabs_wide.plot(kind="bar", ax=ax)
ax.set_title("Liabilities Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(title="Liabilities", loc="upper left", ncol=2, bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()

# ---------- 3. QUARTERLY DATA: denormalize + two stacked charts ----------

q_df = pd.read_excel(
    file_path,
    sheet_name="Data Sheet",
    skiprows=40,      # A41
    nrows=10,
    usecols="A:K"
)

q_long = q_df.melt(
    id_vars=["Report Date"],
    var_name="Date",
    value_name="Amount"
)
q_long["Date"] = pd.to_datetime(q_long["Date"])
q_wide = q_long.pivot_table(
    index="Date",
    columns="Report Date",
    values="Amount",
    aggfunc="sum"
).reset_index()
q_wide["Year"] = q_wide["Date"].dt.year

# 3a. Sales + Other Income + profit-related
components_1 = [
    "Sales",
    "Other Income",
    "Profit before tax",
    "Tax",
    "Net profit",
    "Operating Profit"
]
comp1_cols = [c for c in components_1 if c in q_wide.columns]

fig, ax = plt.subplots(figsize=(12, 6))
q_wide.set_index("Year")[comp1_cols].plot(kind="bar", stacked=True, ax=ax)
ax.set_title("Quarterly Profit Components Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(title="Components", loc="upper left", ncol=2, bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()

# 3b. Expenses + Depreciation + Interest
components_2 = ["Expenses", "Depreciation", "Interest"]
comp2_cols = [c for c in components_2 if c in q_wide.columns]

fig, ax = plt.subplots(figsize=(12, 6))
q_wide.set_index("Year")[comp2_cols].plot(kind="bar", stacked=True, ax=ax)
ax.set_title("Quarterly Expense Components Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(title="Components", loc="upper left", ncol=1, bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()


# ---------- 4. CASH FLOW: denormalize + clustered line chart ----------

cf_df = pd.read_excel(
    file_path,
    sheet_name="Data Sheet",
    skiprows=80,      # A81
    nrows=5,
    usecols="A:K"
)

cf_long = cf_df.melt(
    id_vars=["Report Date"],
    var_name="Date",
    value_name="Amount"
)
cf_long["Date"] = pd.to_datetime(cf_long["Date"])
cf_wide = cf_long.pivot_table(
    index="Date",
    columns="Report Date",
    values="Amount",
    aggfunc="sum"
).reset_index()
cf_wide["Year"] = cf_wide["Date"].dt.year

## Set Year as index and transpose for horizontal clustered bar chart
cf_plot = cf_wide.set_index('Year')[['Cash from Financing Activity', 'Cash from Investing Activity', 'Cash from Operating Activity']]

# Plot clustered column chart with Year on y-axis
ax = cf_plot.plot(kind='barh', figsize=(12, 6))
ax.set_xlabel('Amount')
ax.set_ylabel('Year')
ax.set_title('Cash Flows by Year')
ax.legend(title='Cash Flow Type', loc='upper left', bbox_to_anchor=(1.02, 1.0))
plt.tight_layout()
plt.show()
```
